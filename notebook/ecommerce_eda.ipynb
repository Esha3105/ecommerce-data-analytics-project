The datatype of the date column was object so we have to convert it into datetime format

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
# create a column Age_Group 

labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['Age_Group'] = pd.qcut(df['Age'], q=4, labels = labels)
```

```python
df[['Age', 'Age_Group']].head(10)
```

Created a new column named as Age Group based on Age column

```python
df['Discount_Applied'] = df['Discount_Amount'].apply(lambda x: 'Yes' if x > 0 else 'No')
df.head(10)
```

Create a column based on discount amount whether discount is applied yes or no

```python
df['Year'] = df['Date'].dt.year
df['Month_Name'] = df['Date'].dt.strftime('%B')
df.head(10)
```

extracted month_name, year from date column and created a new column named as month_name and year.

```python
df.to_csv("ecommerce_cleaned.csv", index=False)
```
